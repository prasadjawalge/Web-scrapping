```python
def extract_categories(url):
  categories = {}
  r = requests.get(url)
  soup = BeautifulSoup(r.content)
  categories_list = soup.find("ul",class_="nav nav-list").ul.find_all('li')
  for li in categories_list:
    categories.update({li.text.strip() : url + li.a["href"]})
  return categories
```
